```python
import yfinance as yf
import pandas as pd

import warnings
warnings.filterwarnings('ignore')

# Define the stocks 
stocks = ['AAPL', 'PG', '^GSPC']

# Download all data at once
print("Downloading stock data...")
data = yf.download(stocks, start='2014-01-02', end='2023-12-31')
print("Download complete!")
```

```python
closing_prices = data['Close']
closing_prices.rename(columns={'^GSPC' : 'S&P500'}, inplace=True)
print("\nYour closing prices:")
print(closing_prices.head(10))
```

```python
# Saving the raw data to a csv file
closing_prices.to_csv('../data/raw_data.csv')
```
